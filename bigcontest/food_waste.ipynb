# 연도별 일일 총 쓰레기 배출

## - 목표: 연도를 column으로, 날짜를 row로 하는 쓰레기 배출 dataframe 생성

```python
date_fw = fw.groupby(['base_date'])['em_cnt', 'em_g', 'pay_amt'].agg(['count', 'mean', 'sum'])
# 'mean' = 'sum' / 'count'_배출거점
```

```python
date_fw.head()
```

```python
# date_fw[date_fw.index == '2020-01-01']
```

```python
year_fw_2018 = date_fw[:365]
year_fw_2018
```

```python
year_fw_2019 = date_fw[365:730]
year_fw_2019
```

```python
year_fw_2020 = date_fw[730:1096] # 2020년 366일 (2월 29일)
year_fw_2020
```

```python
year_fw_2021 = date_fw[730:]
year_fw_2021
```

```python
year_fw_2021['em_cnt']['sum']
```

```python
year_fw_2018['2018_em_cnt_sum'] = year_fw_2018['em_cnt']['sum']
year_fw_2018 = year_fw_2018.reset_index(drop=True)

year_fw_2019['2019_em_cnt_sum'] = year_fw_2019['em_cnt']['sum']
year_fw_2019 = year_fw_2019.reset_index(drop=True)

year_fw_2020['2020_em_cnt_sum'] = year_fw_2020['em_cnt']['sum']
year_fw_2020 = year_fw_2020.reset_index(drop=True)

year_fw_2021['2021_em_cnt_sum'] = year_fw_2021['em_cnt']['sum']
year_fw_2021 = year_fw_2021.reset_index(drop=True)
```

```python
year_fw_2018.head()
```

```python
year_fw = pd.DataFrame()
```

```python
# 연도별 일일 count sum
year_fw['2018_cnt'] = year_fw_2018['2018_em_cnt_sum']
year_fw['2019_cnt'] = year_fw_2019['2019_em_cnt_sum']
year_fw['2020_cnt'] = year_fw_2020['2020_em_cnt_sum']
year_fw['2021_cnt'] = year_fw_2021['2021_em_cnt_sum']
year_fw
```

```python
# 연도별 일일 amount sum
year_fw_2018['2018_em_g_sum'] = year_fw_2018['em_g']['sum']
year_fw_2019['2019_em_g_sum'] = year_fw_2019['em_g']['sum']
year_fw_2020['2020_em_g_sum'] = year_fw_2020['em_g']['sum']
year_fw_2021['2021_em_g_sum'] = year_fw_2021['em_g']['sum']
# year_fw_2021 = year_fw_2021.reset_index(drop=True)

year_fw['2018_amt'] = year_fw_2018['2018_em_g_sum']
year_fw['2019_amt'] = year_fw_2019['2019_em_g_sum']
year_fw['2020_amt'] = year_fw_2020['2020_em_g_sum']
year_fw['2021_amt'] = year_fw_2021['2021_em_g_sum']

year_fw
```

```python
# 연도별 일일 amount sum 단위 변환 kg
year_fw['2018_kg'] = year_fw['2018_amt']/1000
year_fw['2019_kg'] = year_fw['2019_amt']/1000
year_fw['2020_kg'] = year_fw['2020_amt']/1000
year_fw['2021_kg'] = year_fw['2021_amt']/1000

year_fw
```

```python
date = date_fw[:365]
date = date.reset_index()
date
```

```python
import datetime

date['base_date'] = pd.to_datetime(date['base_date']) 

# 년도 추출
date['Year'] = date['base_date'].dt.year
# 월 추출
date['Month'] = date['base_date'].dt.month
# 일 추출
date['Day'] = date['base_date'].dt.day

date['Date'] = date['Month'].astype(str) + '월 ' + date['Day'].astype(str) + '일'
date
```

```python
year_fw['Date'] = date['Date']
# year_fw
```
